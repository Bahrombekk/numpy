```python
def merge_image(image1, image2, image3, image4):
    """
    Merges three images into one.
    merge image shape: (700, 700, 3)
    Args:
        image1: The first image.
        image2: The second image.
        image3: The third image.
        image4: The fourth image.
    Returns:
        np.ndarray: The merged image.
    """
    a=np.full((700,700,3),0)
    a[:len(a)//2:,:len(a)//2:]=image1
    a[len(a)//2::,len(a)//2::]=image4
    a[len(a)//2::,:len(a)//2:]=image3
    a[:len(a)//2:,len(a)//2::]=image2
    return a
```

```python
merge_image = merge_image(image1, image2, image3, image4)
plt.imshow(merge_image)
plt.show()

q1.check(merge_image, github)
```
