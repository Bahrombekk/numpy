```python
# import library
import csv
import matplotlib.pyplot as plt
import numpy as np
```

# IMPORT DATA

```python
gender = []
height = []
weight = []
```

```python
def read_csv_data(file_path):
    data = open(file_path)
    data = csv.reader(data)
    next(data)
    """
    Read csv file
    Args:
        file_path(str): file path
    Returns:
        list: data split row
    """
    return list(data)
```

```python
data = read_csv_data('weight-height.csv')
```

```python
print(data[1])
```

```python
def get_data(data):
    """
    Get data from list.
    Gender: Change Male to 0 and Female to 1
    Weight: Place the column in the Kg view given in Pound (1 kg = 2,205 pound).
    Height: Place the column on the list in the cm view given in inches (2.54 cm = 1 inch).
    Args:
        data(list): data split row
    Returns:
        array: gender, weight and height

    """
    gender = []
    weight = []
    height = []
    for i in data:
      weight.append(float(i[2])/2.205)
      height.append(float(i[1])*2.54)
      if i[0]=="Male":
        gender.append(0)
      else:
        gender.append(1)

    return gender,weight,height
    plt.show()
file_path='weight-height.csv'
data = read_csv_data(file_path)

gender,weight,height = get_data(data)
```

```python
data = get_data(data[1:]) #get_data(data[1:10]+data[-10:])
```

```python
plt.scatter(weight,height,c=gender)
plt.show()
```
